# Is there a correlation between the “na_sales” and “jp_sales” for the years 2010-2014?  

```python
filtered_df = df[(df['Year_of_Release'] >= 2010) & (df['Year_of_Release'] <= 2014)]
sales_columns = ['NA_Sales', 'JP_Sales']

correlation_matrix = filtered_df[sales_columns].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues',
            fmt=".2f")

plt.title('Correlation Matrix of NA Sales and JP Sales (2010-2014)')

#plt.savefig('Correlation_matrix.png', dpi=300, bbox_inches='tight')

plt.show()
```

Each cell represents the correlation coefficient between two variables. The color intensity indicates the strength and direction of the correlation: darker colors represent stronger correlations, while lighter colors represent weaker correlations. Additionally, annotations within the cells display the correlation coefficient values.

# What is the distribution of the most popular 4 game genres?  

```python
genre_counts = df['Genre'].value_counts().head(4)

plt.figure(figsize=(10, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='Blues')
plt.title('Top 4 Game Genres')
plt.xlabel('Genre')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y') 

#plt.savefig('top_4_games_genres.png', dpi=300, bbox_inches='tight')

plt.show()
```

# Do older games (2005 and earlier) have a higher MEAN “eu_sales” than newer games (after 2005)?  

```python
older_games = df[df['Year_of_Release'] <= 2005]
newer_games = df[df['Year_of_Release'] > 2005]

mean_eu_sales_older = older_games['EU_Sales'].mean()
mean_eu_sales_newer = newer_games['EU_Sales'].mean()

data = pd.DataFrame({'Release Year': ['Older Games', 'Newer Games'],
                     'Mean EU_Sales': [mean_eu_sales_older, mean_eu_sales_newer]})

plt.figure(figsize=(8, 6))
sns.barplot(x='Release Year', y='Mean EU_Sales', data=data, palette='Blues')
plt.title('Mean EU_Sales of Older vs. Newer Games')
plt.xlabel('Release Year')
plt.ylabel('Mean EU_Sales')
plt.grid(axis='y')  

#plt.savefig('MEAN.png', dpi=300, bbox_inches='tight')

plt.show()
```

# What are the 3 most common “developer” in the dataset?  

```python
col = 'Developer'

chart = df[['Name', col]].groupby([col]).count().sort_values('Name', ascending=False).reset_index()

chart = chart[:3]

sns.set_style("white")
plt.figure(figsize=(12.4, 5))
plt.xticks(rotation=90)
sns.barplot(x=col, y='Name', data=chart, palette='Blues').set_title('Top 3 Developers', fontsize=16)
plt.ylabel('Count (millions)', fontsize=14)
plt.xlabel('')
plt.grid(axis='y')

#plt.savefig('Developer.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
top_developers = chart['Developer'].tolist()

df_top_developers = df[df['Developer'].isin(top_developers)]

developer_counts = df_top_developers['Developer'].value_counts()


plt.figure(figsize=(8, 8))
plt.pie(developer_counts, labels=developer_counts.index, autopct='%1.1f%%', startangle=130, colors=sns.color_palette('Blues'))
plt.title('Distribution of Games by Top 3 Developers', fontsize=14)
plt.axis('equal')  

#plt.savefig('Developer_Pie.png', dpi=300, bbox_inches='tight')

plt.show()
```

# Is there a statistically significant difference in the relative frequency of the top published game based on global sales between the years 2006 and 2016?

```python
top_10_publishers = df.groupby('Publisher')['Global_Sales'].sum().nlargest(10).reset_index()

plt.figure(figsize=(16, 6))
sns.barplot(x='Publisher', y='Global_Sales', data=top_10_publishers, palette='Blues')
plt.xlabel('Publisher')
plt.ylabel('Global Sales (Millions)')
plt.title('Top 10 Publishers by Global Video Games Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 

#plt.savefig('top_10_publishers.png', dpi=300, bbox_inches='tight')

plt.show()

```

```python

ea_games = df[df['Publisher'] == 'Electronic Arts']

ea_games_2006_2016 = ea_games[(ea_games['Year_of_Release'] >= 2006) & (ea_games['Year_of_Release'] <= 2016)]

relative_frequency = ea_games_2006_2016.groupby('Year_of_Release').size() / len(ea_games_2006_2016)


plt.figure(figsize=(10, 6))
relative_frequency.plot(marker='o')
plt.title('Relative Frequency of Electronic Arts Games (2006-2016)')
plt.xlabel('Year')
plt.ylabel('Relative Frequency')
plt.grid(True)

#plt.savefig('Relative_freq.png', dpi=300, bbox_inches='tight')

plt.show()
```

**Why were those specific visualizations chosen to illustrate each answer?**

(Atlassian, n.d.) states that visualisation can help our brains understand complex data. Charts can take large amounts of data and transform it into an understandable format. It is important to note that data can be inserted into any chart type, different charts can better represent your data and use case. 

In this analysis, bar charts were used to illustrate- Top 5 Games by Global Sales, the most popular genres, mean EU sales and the most common developers in the dataset.


These are one the most fundamental data visualisations. Levels are plotted on one chart axis and values are plotted on the other. These are used to show a distribution of data points. We can see which groups are the highest or most common and how different groups compare against others. They are very intuitive to understand, the length of each bar represents the amount. When using a limited number of categories, these are very useful as it presents a concise overview without overwhelming the audience with too much information. Clear labelling was used to identify each point without cluttering up the visualisation. 

A heatmap was also used to visualise a correlation between variables. These 
are very useful as the correlation coefficient ranges from -1 to 1, a value close to 1 indicates a strong positive correlation, while a value close to -1 indicates a strong negative correlation. A value close to 0 indicates no correlation, in other words, no relationship between the variables. The correlation of 0.26 in this graph is weak, suggesting that the relationship between the variables is not very strong. 



To add a little more interest through visualisations, a pie chart was added to show how the Top 3 Developers games were distributed. Pie charts, while not everyones favourite choice, can be beneficial when there are only a few categories to plot. 


The final choice of visualisation was a line plot, this was created to visualise the trend showing the relative frequency of games released by Electronic Arts between 2006 and 2016. Line charts use points connected by line segments from left to right to demonstrate changes in value. They depict a continuous progression across the horizontal axis and values on the vertical axis.
